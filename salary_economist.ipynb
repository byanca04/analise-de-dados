```python
# a media entre o conjunto de dados
media_salario = df['salario'].mean()
media_salario.round(2)
```

```python
# mediana do salário entre o conjunto de dados
df['salario'].median()
```

```python
# salario anual de cada profissional
df['salario_anual'] = df['salario'] * 12
df.head()
```

```python
df['experiencia_anos'] = df['experiencia'].round()
df.head()
```

```python
# apresenta a quantidade de experiencia
df['experiencia_anos'].value_counts()
```

```python
# quais profissões tem experiencia entre 5 e 6 anos
df.loc[(df['experiencia_anos'] >= 5) & (df['experiencia_anos'] <= 6)]
```

```python
# pessoas que ainda não completou um ano de experiência
df.loc[0] # para a seleção e filtragem de dados em DataFrames
df[df['experiencia_anos'] == 0.0]
```

```python
from matplotlib import pyplot as plt

# gráfico de dispersão

plt.scatter(df['experiencia'], df['salario'])
plt.show()
```

```python
# separando a experiencia em estágio, júnio, pleno e sênior
def expert(experiencia_anos):
    if experiencia_anos < 1:
        return "Estágio"
    elif experiencia_anos <= 3:
        return "Júnior"
    elif experiencia_anos <= 5:
        return "Pleno"
    else:
        return "Sênior"
```

```python
df['cargo'] = df['experiencia_anos'].apply(expert) # apply - é uma função de ordem superior, o que significa que ela recebe uma outra função
df.head()
```

```python
# agrupar salario de cada cargo
df.groupby('cargo')['salario'].mean().round(2)
```

```python
plt.bar(df['cargo'], df['salario'])
plt.title("salário por cargo")
plt.xlabel("Salário")
plt.ylabel("Cargo")
plt.show(plt.bar)
```
