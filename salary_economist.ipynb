```python
# a media entre o conjunto de dados
media_salario = df['salario'].mean()
media_salario.round(2)
```

```python
# salario anual de cada profissional
df['salario_anual'] = df['salario'] * 12
df.head()
```

```python
df['experiencia_anos'] = df['experiencia'].round()
df.head()
```

```python
# apresenta a quantidade de experiencia
df['experiencia_anos'].value_counts()
```

```python
from matplotlib import pyplot as plt

# gr√°fico de dispers√£o

plt.scatter(df['experiencia'], df['salario'])
plt.show()
```

```python
# separando a experiencia em est√°gio, j√∫nio, pleno e s√™nior
def expert(experiencia_anos):
    if experiencia_anos < 1:
        return "Est√°gio"
    elif experiencia_anos <= 3:
        return "J√∫nior"
    elif experiencia_anos <= 5:
        return "Pleno"
    else:
        return "S√™nior"
```

```python
df['cargo'] = df['experiencia_anos'].apply(expert) # apply - √© uma fun√ß√£o de ordem superior, o que significa que ela recebe uma outra fun√ß√£o
df.head()
```

```python
# agrupar salario de cada cargo
df.groupby('cargo')['salario'].mean().round(2)
```

```python
plt.bar(df['cargo'], df['salario'])
plt.title("sal√°rio por cargo")
plt.xlabel("Sal√°rio")
plt.ylabel("Cargo")
plt.show(plt.bar)
```
