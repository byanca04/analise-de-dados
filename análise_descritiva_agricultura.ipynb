MAPA DE CALOR


```python
raw_data = ['Coarse wool Price', 'Copra Price', 'Cotton Price','Fine wool Price',
       'Hard log Price', 'Hard sawnwood Price','Hide Price', 'Plywood Price',
       'Rubber Price', 'Softlog Price','Soft sawnwood Price','Wood pulp Price']


corrmat = df[raw_data].corr()

fig = plt.figure(figsize=(12,9))

mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax=0.9, mask=mask, square=True, annot = True)
plt.show(
)
```

O mapa de calor mostra a correlação entre as matérias-primas maior o valor correlacionado maior é a chance de serem duas matérias-primas relacionadas, nmas não necessariamente.
  - quanto mais próximo de 1, maior é a correlação

```python
plt.figure(figsize=(30,15))
changelist=[ 'Coarse wool price % Change', 'Copra price % Change', 'Cotton price % Change',  'Fine wool price % Change',
       'Hard log price % Change','Hard sawnwood price % Change', 'Hide price % change',
        'Plywood price % Change','Rubber price % Change', 'Softlog price % Change',
      'Soft sawnwood price % Change','Wood pulp price % Change']

corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, vmax=0.9, mask=mask, square=True, annot = True)
plt.show()
```

A partir deste gráfico podemos dizer que quase não há relação entra a vairação peercentual dos preços das matérias-primas
  - o valor negativo emplica que duas variáveis estão negativamnte correlacionadas( um aumento, outro descréscimo)
  - zero implica nenhuma relação
  - Caso contrário, maior o valor a change de relação dos preços e seus gráficos de % de mudança

# Fazendo e respondendo perguntas
Q1: Descybra a variação normal do preço de cada matéria-prima

```python
changelist= ['Coarse wool price % Change', 'Copra price % Change', 'Cotton price % Change',  'Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change', 'Hide price % change','Plywood price % Change','Rubber price % Change', 'Softlog price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

for i in range(len(changelist)):
  plt.figure(figsize=(12,12))
  df[changelist[i]].hist(figsize=(11, 9), linewidth=1)
  plt.xlabel('% Change')
  plt.ylabel('count')
  plt.legend(changelist[i:], loc='upper center', bbox_to_anchor=(1.2,1))
```

Q2 - Encontre a matéria prima que tem o menor preço ao longo dos anos

```python
plt.figure(figsize=(10, 10))
materialsList = ['Coarse wool Price', 'Copra Price', 'Cotton Price','Fine wool Price',
       'Hard log Price', 'Hard sawnwood Price','Hide Price', 'Plywood Price',
       'Rubber Price', 'Softlog Price','Soft sawnwood Price','Wood pulp Price']

for i in range(len(materialsList)):
  plt.subplot(4, 3, i+1)
  plt.subplots_adjust( hspace=0.5, wspace=0.5)
  plt.plot(df[materialsList[i]])
  plt.title(materialsList[i])
  plt.xticks(rotation=90)
plt.suptitle("Raw-Master price comparision")
```

```python
plt.figure(figsize=(10,10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title("Raw-Material price comparasion")
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(['Cotton Price', 'Rubber Price'], loc='upper center', bbox_to_anchor=(1.2,1))
```


Q3: qual matéria-prima tem a maior e menor variação de % de preço


```python
import random
plt.figure(figsize=(12,12))
for i in range(len(changelist)):

  r = random.random()
  b = random.random()
  g = random.random()
  color = (r, g, b)
  plt.subplot(4, 3, i+1)
  plt.subplots_adjust( hspace=0.5, wspace=0.5)
  plt.plot(df[changelist[i]], color=color)
  plt.title(changelist[i])
  plt.xticks(rotation=90)
  plt.subplot(4, 3, i+1)
  plt.subplots_adjust( hspace=0.5, wspace=0.5)
```
