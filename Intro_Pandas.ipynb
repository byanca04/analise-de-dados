## DataFrame & Series

- `Series`: é uma **coluna** de dados. Pode contar diferentes tipos de dados, mas eles não podem se misturar dentro da mesma coluna.
  * ela é unidimensional, significa que representa apenas uma coluna;
- `DataFrame`: é uma **tabela** de dados. Conjunto de `Series`.
  * ele é bidimensional, significa que contém linha e coluna;

## Métodos `loc` & `iloc`

- `loc`: acessa pelo rótulo (label) - nome da coluna ou nome do índice
- `iloc`: acesso pelo índice da coluna ou linha

```python
df.head()

# acessando a linha 0 da coluna 'produto'
df.loc[0, ['produto', 'vendedor']]
```

```python
# linhas 4 e coluna 1
df.iloc[4, 1]
```

```python
# linhas a partir do 2 até o 6 da coluna 2
df.iloc[2:6, 2]
```

```python
# escolhendo as linhas 0, 1, e 5 das colunas produto e vendedor
df.loc[[0, 1, 5], ["produto", "vendedor"]]
```

## Inserindo e excluindo colunas

## Contagem de valores distintos

- Métodos `value_counts()`, `unique()`, `nunique()`

```python
# apresenta a quantidade de cada produto da coluna 'produto'
df['produto'].value_counts()
```

```python
# lista os produtos da coluna 'produto'
df['produto'].unique()
```

```python
# mostra quantos produtos tem na coluna 'produto'
df['produto'].nunique()
```

## Agrupando dados

- Método `groupby()` e `pivot_table()`

```python
df.groupby('produto').sum()
```

```python
df.groupby('vendedor')['quantidade'].sum()
```

```python
df.pivot_table(
    index='produto',
    columns='vendedor',
    values='quantidade',
    aggfunc='sum',
    fill_value= 0,
    margins=True # a soma total das linhas e colunas
)
```

## Filtrando linhas

- Método `query()`

```python
df.query('quantidade < 2')
```

```python
df.query("produto == 'Tablet'")
```
