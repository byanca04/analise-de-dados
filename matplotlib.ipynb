```python
import matplotlib.pyplot as plt
```

```python
# explorando o matplotlib de linhas
plt.plot([1, 2, 3, 4], [1, 5, 2, 6], color="red", linestyle="--", marker="o", label="linha vermelho")
plt.plot([1, 2, 3, 4], [2, 2, 2, 2], color="green", linestyle="-", label="linha verde")

plt.title("camarada montanha")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.legend()

plt.show()
```

```python
plt.savefig("figura-montanha.png")
```

```python
# matplotlib de barras

categorias = ["A", "B", "C", "D"]
valores = [5, 6, 9, 4]

plt.barh(categorias, valores)
plt.title("gráfico de barra")
plt.show()

plt.bar(categorias, valores)
plt.title("segundo gráfico de barra")
plt.show(plt.bar)
```

```python
# matplotlib histograma
# gráfico de barras que demonstra uma distribuição de frequências
dados = [1, 4, 3, 2, 4, 5, 2, 6, 7, 9, 7, 9, 10, 5]

plt.hist(dados, bins=30, edgecolor="black")
plt.show()
```

```python
# matplotlib, gráfico de dispersão

x = [20, 3, 5, 17, 14, 12, 5, 9, 22, 13]
y = [30, 20, 14, 34, 55, 42, 39, 100, 68, 30]

plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# matplotlib, gáfico de pizza

fatias = [ 40, 50, 65]
frutas = [ "Melancia", "Lichia", "Abacaxi"]

plt.pie(fatias, labels=frutas, autopct="%1.2f%%" ,startangle=120)

plt.style.use('classic')
plt.show()
```
